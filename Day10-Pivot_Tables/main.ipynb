Here we will look at `pivto` and `pivot tables` in Pandas.

Pivot: Allows us to reshape or transform our dataframe. If we want to have the `cities` on the x-axis and `date` on the y-axis for the following table, we can use Pivot.

```python
df = pd.read_csv('weather.csv')
df
```

```python
df.pivot(index='date', columns='city', values = 'humidity')
```

Now if we want to have just a particular value, we can pass it as an argument to `values` parameter.

```python
df.pivot(index='date', columns='city', values = 'humidity')
```

Now, pivot table allows one to summarize and aggregate data inside a dataframe.

```python
df = pd.read_csv('weather2.csv')
df
```

```python
df.pivot_table(index = 'city', columns = 'date', aggfunc = 'sum')
```

```python
# now we will look at grouper, which is aggregate the data based on date frequency 
df = pd.read_csv('weather3.csv')
df['date'] = pd.to_datetime(df['date'])
df.pivot_table(index = pd.Grouper(freq='M', key = 'date'), columns = 'city')
```
