# Pandas Tutorial Day 12

We will look at one more dataframe reshaping technique called `stack-unstack`.

```python
import pandas as pd

df = pd.read_excel('stocks.xlsx', header=[0,1])
df
```

```python
df_stack = df.stack()
df_stack
```

```python
df_stack.unstack()
```

```python
# using stack with 3 level header file
df2 = pd.read_excel('stocks_3_levels.xlsx', header = [0,1,2])
df2
```

```python
df2.stack(future_stack=True, level=2)
```
