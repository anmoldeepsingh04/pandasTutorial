We will look at different ways to create pandas dataframes. We will look at the following methods:
1. Using CSV and Excel
2. From python dictionary
3. From list of tuples

```python
import pandas as pd
```

## Using CSV and Excel

```python
df = pd.read_csv('database.csv')
df

# similarly, we can have a .xlsx file and we can create a dataframe using the following:
# df_excel = pd.read_excel('file.xlsx', 'Sheet1')
```

## Using Python Dcitionary

```python
dataDict = {
    'ID':['1', '2', '3', '4', '5'],
    'Name':['Ajay', 'Ramesh', 'Kris', 'Alto', 'Tenny'],
    'Surname': ['Singh', 'Kumar', 'Grusing', 'Pat', 'Jennings'],
    'Score': [22, 37, 50, 72, 10],
    'Age':[26, 22, 31, 30, 28]
}
df = pd.DataFrame(dataDict)
df
```

## From List of Tuples

```python
dataTup = [
    ('1', '2', '3', '4', '5'),
    ('Ajay', 'Ramesh', 'Kris', 'Alto', 'Tenny'),
    ('Singh', 'Kumar', 'Grusing', 'Pat', 'Jennings'),
    (22, 37, 50, 72, 10),
    (26, 22, 31, 30, 28)
]
df = pd.DataFrame(dataTup, columns=['ID', 'Name', 'Surname', 'Score', 'Age'])

lis = []
for cols in df.columns:
    lis.append(df[cols])

liss = []
for i in range(len(lis)):
    liss.append(tuple(lis[i]))

df = pd.DataFrame(liss, columns=[['ID', 'Name', 'Surname', 'Score', 'Age']])
df
```
