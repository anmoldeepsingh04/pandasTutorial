```python
import pandas as pd
import numpy as np
```

We will see how we can replace certain values that are not useful for us

```python
df = pd.read_csv('weather_data.csv')
# replacing -99999 with NaN
# new_df = df.replace([-99999, -88888], np.nan)

# if we want to replace values based on specific columns, we can use a dictionary
# new_df = df.replace({
#     'temperature': -99999,
#     'windspeed' : -99999,
#     'event' : '0'
# }, np.nan)

# if we want to create a mapping to replace each entry with something else no matter the column, we can use a dictionary
new_df = df.replace({
    -99999 : np.nan,
    "0" : "Sunny"
})
new_df
```

Now if our data has some unneccesarry text along with the data in the same cell (eg. 32F, 45C, for temperatures), we can use regular expressions to deal with those values and remove them

```python
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed':'[A-Za-z]'}, "", regex=True)

new_df
```

Now we'll see how to replace a list of values with another list of values

```python
df = pd.DataFrame({
    'score' : ['exceptional', 'average','good','poor','average','exceptional'],
    'student' : ['rob', 'maya','pativ', 'tom','julian','erica']
})
num = [4,2,3,1]

# replacing score list with numerical values
df = df.replace(list(set(df['score'])), [4,2,3,1])

df
```
