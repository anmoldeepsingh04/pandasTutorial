We have the temperature data for some cities and we want to answer some questions:
1. Find the maximum temperature in each of the cities
2. Find average wind speed per city

So we are grouping the data per city and then find some analytics related to it. This strategy is called SPLIT->APPLY->COMBINE

```python
import pandas as pd
df = pd.read_csv('weather_by_cities.csv')
df
```

## groupby()

```python
# this gives a dataframegroupby object, where the city is the key and rest of its data is the value
g = df.groupby('city')

# to display each dataframe corresponding to city
for city, city_df in g:
    print(city)
    print(city_df)

# to access particular city data
g.get_group('mumbai')
```

```python
# to find the maximum temperature for each city
g.max()
```

```python
# average speed in each city
g.mean(numeric_only = True)
```

```python
# all the statistics in one go
g.describe()
```

```python
%matplotlib inline
g.plot()
```
